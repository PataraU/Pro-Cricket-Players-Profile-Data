```python
df = pd.read_csv('data for cleaning.csv')
```

```python
#renmae the columns
```

```python
df = df.rename(columns = {'NO' : 'Not_Outs', 'HS' : 'High_Inns_Scores',  'BF' : 'Balls_Faced', 'SR' : 'Strike_rates'})
```

```python
#Find a null value
```

```python
df.isnull().sum()
```

```python
df[df['Balls_Faced'].isna()  == 1]
```

```python
df['Balls_Faced'] =  df['Balls_Faced'].fillna(0)
```

```python
df
```

```python
#Drop and remove the duplicates
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated()
```

```python
df
```

```python
#Split starting and ending years
```

```python
df['Span'].str.split(pat = '-')
```

```python
df.head(10)
```

```python
df['Start_Year'] = df['Span'].str.split(pat = '-').str[0]
```

```python
df['Final_Year'] = df['Span'].str.split(pat = '-').str[1]
```

```python
df
```

```python
df = df.drop(['Span'], axis = 1)
```

```python
#Split the nationality of the players
```

```python
df['Player'].str.split(pat = '(') 
```

```python
df['Nationality'] = df['Player'].str.split(pat = '(').str[1]
```

```python
df.head(5)
```

```python
df['Nationality'] = df['Nationality'].str.rstrip(')')
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
```

```python
df
```

```python
#Change data types as High_Inns_Scores should be an int
```

```python
df.dtypes
```

```python
df.head(10)
```

```python
df['High_Inns_Scores'] = df['High_Inns_Scores'].str.rstrip('*')
```

```python
df['High_Inns_Scores'] = df['High_Inns_Scores'].astype('int')
```

```python
df['4s'] = df['4s'].str.rstrip('+')
```

```python
df['Balls_Faced'] = df['Balls_Faced'].str.rstrip('+')
```

```python
df['Balls_Faced'] = df['Balls_Faced'].replace('-', 0)
```

```python
df.head(15)
```

```python
df['Balls_Faced'] = df['Balls_Faced'].astype('float')
```

```python
df.head(10)
```

```python
df = df.astype({'Start_Year' : 'int', 'Final_Year' : 'int'})
```

```python
df.dtypes
```

```python
 ###Question what are the career length of each player
```

```python
df['Career_Length'] = df['Final_Year'] - df['Start_Year']
```

```python
df
```

```python
##Question 1: what is the average career length of all players
```

```python
df['Career_Length'].mean().round(2)
```

```python
##Question 2: What is the average Strike_rates of players who played over 10 years
```

```python
Over_Decade = df[df['Career_Length'] > 10]
```

```python
Over_Decade['Strike_rates'].mean()
```

```python
##Question 3: What is the number of players who played before 1960
```

```python
df[df['Start_Year'] < 1960]['Player'].count()
```

```python
##Question 4: What country has the highest "High_Inns_Scores"?
```

```python
Highest_Inns_Scores = df.groupby('Nationality')['High_Inns_Scores'].max().sort_values(ascending=False)
```

```python
Highest_Inns_Scores.head(1)
```

```python
##Plotting Graph 1: Top 10 players who have the highest High_Inns_Scores
```

```python
Top_10_Players = df[['Player', 'High_Inns_Scores']].head(10).sort_values('High_Inns_Scores', ascending = False)
```

```python
Top_10_Players
```

```python
Top_10_Players['High_Inns_Scores'].mean()
```

```python
df['High_Inns_Scores'].mean()
```

```python
#visualize using a bar plot
ax = sns.barplot(x='High_Inns_Scores', y='Player', data=Top_10_Players, palette='viridis')

#add value labels to each player's high_Inns_scores
for index, value in enumerate(Top_10_Players['High_Inns_Scores']):
    plt.text(value + 1, index, f'{value}', va='center')
    
#add
plt.title("Top 10 players who have the highest High_Inns_Scores")

#Leghthen the size of x 
plt.xlim(0, Top_10_Players['High_Inns_Scores'].max() + 50) 

plt.show()
```
